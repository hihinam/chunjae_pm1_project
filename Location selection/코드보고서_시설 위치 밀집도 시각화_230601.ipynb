```python
import folium
import json
import pandas as pd
from folium.plugins import HeatMap
```

```python
geo = json.load(open('./HangJeongDong_ver20230401.geojson.txt', encoding='utf8'))
```

```python
features = []
```

```python
for i in range(len(geo['features'])):
    if geo['features'][i]['properties']['sggnm'] == '화성시':
        features.append(geo['features'][i])
```

```python
geo['features'] = features
```

```python
df = pd.read_excel('./화성시 초중 위도 경도 데이터.xlsx')
```

```python
df.head()
```

```python
point = [[df.loc[i]['Latitude'], df.loc[i]['Longitude']] for i in df.index]
```

```python
title = '화성시 초등학교 밀도'
title_html = f'<h3 style="font-size:16px"><b>{title}</b></h3>' 

m = folium.Map(
    location=[37.2,126.83],
    zoom_start=10, 
    width=640,
    height=480,
)

folium.GeoJson(
    geo,
    name='행정동명'
).add_to(m)

for i in df.index:
    folium.Circle(
        location = [df.loc[i]['Latitude'],df.loc[i]['Longitude']],
        radius = 50,
        color = '#000000',
        fill = 'crimson',
    ).add_to(m)
```
